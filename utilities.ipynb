```python
import pandas as pd
```

```python
column_names = ['id','price_paid', 'transaction_date', 'postcode',
           '',
           '',
           '',
           'building_name_or_number',
           '',
           'street','locality', 'town','district','county',
           '',
          '']
```

```python
#data = pd.read_csv('data/pp-complete.csv', sep=',') #main file with all year
df = pd.read_csv('data/pp-2019.csv', sep=',', header=None) #temp 2019-only file for faster run locally
```

```python
df.columns = ['id',
                'price_paid',
                'transaction_date',
                'postcode',
               'property_type',
               'new_build',
               'estate_type',
               'building_name_or_number',
               'secondary_name',
               'street',
                'locality',
                'town',
                'district',
                'county',
           'unk_1',
          'unk_2']
```

```python
df.head()
```

```python
df.info()
```

```python
#kill id column
df.drop('id', axis=1, inplace=True)
df.head()
```

```python
#cast objects to proper datatypes
df.postcode = df.postcode.astype('string')
df.property_type = df.property_type.astype('string')
df.new_build = df.new_build.astype('bool')
df.estate_type = df.estate_type.astype('string')
df.building_name_or_number = df.building_name_or_number.astype('string')
df.secondary_name = df.secondary_name.astype('string')
df.street = df.street.astype('string')
df.locality = df.locality.astype('string')
df.town = df.town.astype('string')
df.district = df.district.astype('string')
df.county = df.county.astype('string')
df.property_type = df.property_type.astype('string')
df.unk_1 = df.unk_1.astype('string')
df.unk_2 = df.unk_2.astype('string')
df.transaction_date = pd.to_datetime(df['transaction_date'], format='%Y-%m-%d')

df.info()
```
